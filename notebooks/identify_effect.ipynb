```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loyal_customers = pd.read_csv('../data/loyal_customer_data.csv')
```

```python
lc_period34 = loyal_customers[(loyal_customers['period_code'] == 3) | (loyal_customers['period_code'] ==4)]
```

```python
lc_period4 = loyal_customers[loyal_customers['period_code'] ==4]
lc_period3 = loyal_customers[loyal_customers['period_code'] ==3]
```

```python
lc_period12 = loyal_customers[(loyal_customers['period_code'] == 1) | (loyal_customers['period_code'] ==2)]
```

```python
lc_period34
```

```python
lc_34_with_lp = lc_period34[lc_period34['redeemer_latest_ind'] == 1]
lc_34_without_lp = lc_period34[lc_period34['redeemer_latest_ind'] == 0]
```

```python
lc_4_with_lp = lc_period4[lc_period4['redeemer_latest_ind'] == 1]
lc_4_without_lp = lc_period4[lc_period4['redeemer_latest_ind'] == 0]
```

```python
average_purchase_value0 = lc_period3.groupby('date_code')['revenue_after_discount_incl_vat'].mean()
average_purchase_value4 = lc_period4.groupby('date_code')['revenue_after_discount_incl_vat'].mean()


average_purchase_value1 = lc_4_with_lp.groupby('date_code')['revenue_after_discount_incl_vat'].mean()
average_purchase_value2 = lc_4_without_lp.groupby('date_code')['revenue_after_discount_incl_vat'].mean()

plt.plot(average_purchase_value0)
plt.plot(average_purchase_value1)
plt.plot(average_purchase_value2)
plt.xlabel('Date')
plt.ylabel('Average Purchase Value')
plt.title('Average Purchase Value Over Time')
period_4_code = loyal_customers[loyal_customers['period_code'] == 4]['date_code'].min()
plt.axvline(x=period_4_code, color='r')
plt.show()
```
